10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.



```python
# Create the total_bedroom_size column
housing['total_bedroom_size'] = housing['total_bedrooms'].apply(lambda x: 'small' if x <= 10 else 'medium' if 10 < x < 1000 else 'large')#The lambda function checks if the total number of bedrooms is 10 or less, in which case the size is "small". If the number of bedrooms is greater than 10 but less than 1000, the size is "medium". If the number of bedrooms is greater than or equal to 1000, the size is "large".
housing
```

from the above we can see the added  new column named "total_bedroom_size" that categorizes each row based on the number of total bedrooms. 

   If the total bedroom size is <=10, it should be quoted as "small". If the total bedroom size is >=11 or <=1000, it should be "medium",  if total bed room size is >1000  it is considered as "large".
